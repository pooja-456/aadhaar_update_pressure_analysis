```python
import pandas as pd
import os
```

```python
enrolment_agg_df=pd.read_csv("../final_dataset/enrolment_aggregated.csv",parse_dates=["date"])
```

```python
demographic_agg_df=pd.read_csv("../final_dataset/demographic_aggregated.csv",parse_dates=["date"])
```

```python
enrolment_agg_df["year_month"]=(enrolment_agg_df["date"].dt.to_period("M").astype(str))
```

```python
demographic_agg_df["year_month"]=(
  demographic_agg_df["date"].dt.to_period("M").astype(str)
)
```

```python
enrolment_monthly_df=(
  enrolment_agg_df.groupby(["year_month","state_final","district_norm"],as_index=False).agg({"age_0_5":"sum","age_5_17":"sum","age_18_greater":"sum","total_enrolments":"sum"})
)
```

```python
demographic_monthly_df=(demographic_agg_df.groupby(["year_month","state_final","district_norm"],as_index=False).agg({"demo_age_5_17":"sum","demo_age_17_":"sum","total_demographic_updates":"sum"}))
```

```python
print("Original enrolments:",enrolment_agg_df["total_enrolments"].sum())
```

```python
print("Monthly enrolments:",enrolment_monthly_df["total_enrolments"].sum())
```

```python
(enrolment_monthly_df["total_enrolments"]>=0).all
```

```python
print("Original demographic updates:",demographic_agg_df["total_demographic_updates"].sum())
```

```python
print("Monthly demographic updates:",demographic_monthly_df["total_demographic_updates"].sum())
```

```python
(demographic_monthly_df["total_demographic_updates"]>=0).all
```
