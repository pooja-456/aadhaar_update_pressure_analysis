```python
import pandas as pd
import os
```

```python
ENROLMENT_RAW_PATH="../raw_dataset/api_data_aadhar_enrolment"
```

```python
enrolment_files=[ os.path.join(ENROLMENT_RAW_PATH,file)for file in os.listdir(ENROLMENT_RAW_PATH)if file.endswith(".csv")]
```

```python
print(f"Number of enrolment files found: {len(enrolment_files)}")
```

```python
enrolment_df_list=[]
for file in enrolment_files:
  df=pd.read_csv(file)
  enrolment_df_list.append(df)
enrolment_raw_df=pd.concat(enrolment_df_list,ignore_index=True)
```

```python
print("Enrolment Raw Data Shape:",enrolment_raw_df.shape)
print("\nColumns:")
print(enrolment_raw_df.columns.tolist())
print("\nFirst 5 rows:")
enrolment_raw_df.head()
```

```python
parsed_dayfirst=pd.to_datetime(enrolment_raw_df["date"],dayfirst=True,errors="coerce")
parsed_monthfirst=pd.to_datetime(enrolment_raw_df["date"],dayfirst=False,errors="coerce")
print("Invalid dates (dayfirst=True):",parsed_dayfirst.isna().sum())
print("Invalid dates (dayfirst=False):",parsed_monthfirst.isna().sum())
```

```python
enrolment_clean_df=enrolment_raw_df.copy()
enrolment_clean_df["date"]=enrolment_clean_df["date"].str.strip()
enrolment_clean_df["date"]=pd.to_datetime( enrolment_clean_df["date"],dayfirst=True,errors="coerce")
print("Final invalid dates:",enrolment_clean_df["date"].isna().sum())
enrolment_clean_df["date"].head()
```

```python
enrolment_clean_df["state_clean"]=(enrolment_clean_df["state"].str.lower().str.strip().str.replace("&","and",regex=False).str.replace(r"\s+"," ",regex=True))
```

```python
canonical_states=sorted(["andhra pradesh","arunachal pradesh","assam","bihar","chhattisgarh","goa","gujarat","haryana","himachal pradesh","jharkhand","karnataka","kerala","madhya pradesh","maharashtra","manipur","meghalaya","mizoram","nagaland","odisha","punjab","rajasthan","sikkim","tamil nadu","telangana","tripura","uttar pradesh","uttarakhand","west bengal","andaman and nicobar islands","chandigarh","dadra and nagar haveli and daman and diu","delhi","jammu and kashmir","ladakh","lakshadweep","puducherry"])
```

```python
dataset_states=sorted(enrolment_clean_df["state_clean"].unique())
extra_states=sorted(set(dataset_states)-set(canonical_states))
missing_states=sorted(set(canonical_states)-set(dataset_states))
print("Extra/problematic states in dataset:")
extra_states
print("\nCanonical states missing from dataset:")
missing_states
```

```python
state_mapping={
       "orissa":"odisha",
       "pondicherry":"puducherry",
       "dadra and nagar haveli":"dadra and nagar haveli and daman and diu",
       "daman and diu":"dadra and nagar haveli and daman and diu",
       "the dadra and nagar haveli and daman and diu":"dadra and nagar haveli and daman and diu",
       "west bangal":"west bengal",
       "westbengal":"west bengal"
}
```

```python
enrolment_clean_df["state_mapped"]=(enrolment_clean_df["state_clean"].replace(state_mapping))
```

```python
enrolment_clean_df["state_final"]=enrolment_clean_df["state_mapped"].where(enrolment_clean_df["state_mapped"].isin(canonical_states),"Unknown/Invalid")
```

```python
enrolment_agg_df=(enrolment_clean_df.groupby(["date","state_final","district_norm"],as_index=False).agg({"age_0_5":"sum","age_5_17":"sum","age_18_greater":"sum"}))
```

```python
enrolment_agg_df["total_enrolments"]=(enrolment_agg_df["age_0_5"]+enrolment_agg_df["age_5_17"]+enrolment_agg_df["age_18_greater"])
```

```python
print("Aggregated shape:",enrolment_agg_df.shape)
enrolment_agg_df.head()
```

```python
print("original total enrolments:",enrolment_clean_df[["age_0_5","age_5_17","age_18_greater"]].sum().sum())
print("aggregated total enrolments:",enrolment_agg_df["total_enrolments"].sum())
```

```python
district_lookup=(enrolment_clean_df.groupby(["state_final","district_norm"])["district"].agg(lambda x:x.mode().iloc[0]).reset_index())
enrolment_agg_df=enrolment_agg_df.merge(district_lookup,on=["state_final","district_norm"],how="left")
```
